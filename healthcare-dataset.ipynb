```python
import pandas as pd
import numpy as np 

df['Date of Admission'] = pd.to_datetime(df['Date of Admission'])
df['Discharge Date'] = pd.to_datetime(df['Discharge Date'])

total_records = df.shape[0]
avg_stay = df['Recovery_Days'].mean() if 'Recovery_Days' in df.columns else (df['Discharge Date'] - df['Date of Admission']).dt.days.mean()
total_revenue = df['Billing Amount'].sum()
avg_bill = df['Billing Amount'].mean()

print("Executive Metrices Calculated Successfully!\n")
print(f"Total Patients Tracked : {total_records:,}")
print(f"Avg Recovery/Stay Days : {avg_stay:.2f} Days")
print(f"Total Hospital Revenue : ${total_revenue:,.2f}")
print(f"Average Bill Amount    : ${avg_bill:,.2f}")
```

```python
condition_analysis = df.groupby('Medical Condition').agg(
    Patient_Count=('Billing Amount', 'count'),
    Average_Bill=('Billing Amount', 'mean'),
    Average_Stay=('Discharge Date', lambda x:(x - df.loc[x.index, 'Date of Admission']).dt.days.mean())
).reset_index()

print("Disease-wise Analysis")
print(condition_analysis.to_string(index=False))
```

```python
formatted_analysis = condition_analysis = df.groupby('Medical Condition').mean(numeric_only=True) 

if 'Average_Bill' in formatted_analysis.columns:
    formatted_analysis['Average_Bill'] = formatted_analysis['Average_Bill'].map('${:,.2f}'.format)

if 'Average_Stay' in formatted_analysis.columns:
    formatted_analysis['Average_Stay'] = formatted_analysis['Average_Stay'].map('{:.1f} Days'.format)

count_col = [col for col in formatted_analysis.columns if col not in ['Medical Condition', 'Average_Bill', 'Average_Stay']]
if count_col:
    formatted_analysis[count_col[0]] = formatted_analysis[count_col[0]].map('{:,}'.format)

print("Clean Disease-wise Analysis")
print(formatted_analysis.to_string(index=False))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))

sns.barplot(x=condition_analysis.index, y='Billing Amount', data=condition_analysis, palette='viridis')

plt.title('Average Biling Amount by Medical Condtion', fontsize=14, fontweight='bold')
plt.xlabel('Medical Condition', fontsize=12)
plt.ylabel('Average Billing Amount ($)', fontsize=12)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

le = LabelEncoder()
df_ml = df.copy()

df_ml['Gender'] = le.fit_transform(df_ml['Gender'])
df_ml['Blood Type'] = le.fit_transform(df_ml['Blood Type'])

x = df_ml[['Age', 'Gender', 'Blood Type']]
y = le.fit_transform(df_ml['Medical Condition'])

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(x_train, y_train)

y_pred = model.predict(x_test)
print(f"Model Training Complete! Accuracy Score: {accuracy_score(y_test, y_pred):.2%}")
```
